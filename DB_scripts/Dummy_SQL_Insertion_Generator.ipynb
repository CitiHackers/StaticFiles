```python
import string
import random
import numpy as np
```

```python
# General formation
formattedString = "INSERT INTO HOUSE (NAME, PRICE, IS_PUBLIC, LATITUDE, LONGITUDE) VALUES ("
left_latitude = 1.26
right_latitude = 1.31
up_longitude = 103.80
down_longitude = 103.83
```

```python
# random generator
def name_generator(size=10, chars=string.ascii_uppercase + string.ascii_lowercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))
def latitude_generator():
    return round(random.gauss(1.285, 0.012),10)
def longitude_generator():
    return round(random.gauss(103.815, 0.008),10)
def price_generator():
    x = round(random.gauss(600000, 300000),10)
    while (x <= 0):
        x = round(random.gauss(600000, 300000),10)
    return x
```

```python
# script generator
def generate():
    return formattedString + "\"" + \
            name_generator() + "\"," + \
            str(price_generator()) + ", true, " + \
            str(latitude_generator()) + ", " + str(longitude_generator()) + ");"
```

```python
def get_scripts(size):
    for i in range(size):
        print(generate())
```
